```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
laliga_data = pd.read_csv('laliga_CurrentRank.csv')

# Selecting predictors and the target variable
predictors = ['mp', 'w', 'd', 'l', 'gf', 'ga', 'gd', 'xg', 'xga', 'xgd']
target = 'pts'

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(laliga_data[predictors], laliga_data[target], test_size=0.2, random_state=0)

# Creating a linear regression model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)

# Making predictions
predictions = model.predict(X_test)

# Evaluating the model
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print("Root Mean Squared Error: ", rmse)

# Plotting the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='gf', y='pts', data=laliga_data, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Goals For vs Total Points')  # Title for the plot
plt.xlabel('Goals For (gf)')  # Label for the x-axis
plt.ylabel('Total Points (pts)')  # Label for the y-axis
plt.show()
```
